```python
print("Gender Classifier Using Decision Tree by Kal El")

from sklearn import tree #lets you build a decision tree
```

```python
# [height, weight, shoe_size]
X = [[181, 80, 44], [177, 70, 43], [160, 60, 38], [154, 54, 37], [166, 65, 40],
     [190, 90, 47], [175, 64, 39],
     [177, 70, 40], [159, 55, 37], [171, 75, 42], [181, 85, 43]]

Y = ['male', 'male', 'female', 'female', 'male', 'male', 'female', 'female',
     'female', 'male', 'male']
```

```python
#variable to store the decision tree model
clf=tree.DecisionTreeClassifier()

#training

clf=clf.fit(X,Y) #fit method trains the decion tree on our dataset

#variable to store the result of the prediction

prediction=clf.predict([[190,70,43]])

print(prediction)
```
